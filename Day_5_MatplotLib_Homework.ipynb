# Matplotlib

```
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```
# Using numpy's linspace to create linearly spaced numbers

np.linspace(10,20,9)
```

```
import matplotlib.pyplot as plt
# this cuts down the clutter so we dont have to type pypolt everytime
plt.plot([11, 1, 2, 3, 4])
plt.ylabel('some numbers')
plt.show()
```

```
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')
# this sents the range of the axis
plt.axis([0, 6, 0, 20])
plt.show()
```

```
# evenly sampled points
t = np.arange(0., 5., 0.2)

# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', label='t')
plt.plot(t, t**2, 'bs', label='t^2')
plt.plot(t, t**3, 'g^', label='t^3')
plt.title('Plot')
plt.legend(loc="upper left")

plt.show()
```

```
plt.style.use('seaborn-whitegrid')
##sub functions for the figure
fig = plt.figure()
## Axis values
ax = plt.axes()
```

```
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x));
```

```
# Alternatively, we can use the pylab interface

plt.plot(x, np.sin(x));
```

```
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x));
```

```
plt.plot(x, x + 0, '-g')  # solid green
plt.plot(x, x + 1, '--c') # dashed cyan
plt.plot(x, x + 2, '-.k') # dashdot black
plt.plot(x, x + 3, ':r');  # dotted red
```

```
plt.plot(x, np.sin(x))

plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5);
```

```
plt.plot(x, np.sin(x))
plt.axis('tight');
```

```
plt.plot(x, np.sin(x))
plt.axis('equal');
```

```
# Save your plot as a jpg file

plt.plot(x, np.sin(x))
plt.axis('equal');
plt.savefig('plot.jpg')
```

```
# label plot

plt.plot(x, np.sin(x))
plt.title("A Sine Curve")
plt.xlabel("x")
plt.ylabel("sin(x)");
```

```
# Histograms

plt.style.use('seaborn-white')

data = np.random.randn(1000)
plt.hist(data);
```

```
#The hist() function has other options 

plt.hist(data, bins=30, alpha=0.5, color='steelblue');
```

```
plt.subplots(2,1)

plt.subplot(2,1,1)
plt.text(0.5, 0.5, str((2, 1, 1)));

plt.subplot(2,1,2)
plt.text(0.5, 0.5, str((2, 1, 2)));
```

```
plt.subplots(2,2, figsize=(10,10))

plt.subplot(2,2,1)
plt.text(0.5, 0.5, str((2, 1, 1)), fontsize=15, ha='center');

plt.subplot(2,2,2)
plt.text(0.5, 0.5, str((2, 1, 2)), fontsize=15, ha='center');

plt.subplot(2,2,3)
plt.text(0.5, 0.5, str((2, 1, 3)), fontsize=15, ha='center');

plt.subplot(2,2,4)
plt.text(0.5, 0.5, str((2, 1, 4)), fontsize=15, ha='center');
```

```
plt.subplots(2,3, figsize=(10,5))

for i in range(1, 7):
    plt.subplot(2, 3, i)
    plt.text(0.5, 0.5, str((2, 3, i)),
             fontsize=15
             , ha='center')
```
